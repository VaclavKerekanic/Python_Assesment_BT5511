```python
data_frames = []

for i, file in enumerate(file_list):
    if i == 0:  ## specifies the first file in the list
        df = pd.read_table(file)
        column_names = df.columns  ## store the column names from the first file
    else:
        df = pd.read_table(file, skiprows=1, header=None, names=column_names) ## skipping the first row for all the subsequent files and uses the stored column names
    
    data_frames.append(df)

combined_df = pd.concat(data_frames, ignore_index=True) ##combines multiple panda dfs vertically

```

```python
## function for defining the mutation classifications
def mutation_class(row):
    
    wildtype_seq = row['WildType.Sequence']
    mutant_seq = row['Mutant.Sequence']
    
    ## no mutation
    if wildtype_seq == mutant_seq:
        return 'No Mutation'
    
    ## substitution = difference in composition but not length
    if len(wildtype_seq) == len(mutant_seq):
        return 'Substitution' 
    ## insertion = change in legth - longer
    elif len(wildtype_seq) < len(mutant_seq):
        return 'Insertion'  
    ## deletion = change in length - shorter    
    elif len(wildtype_seq) > len(mutant_seq):
        return 'Deletion'  
    
    ## in case of any other mutation types not mentioned above = defensive programming
    return 'Other'

combined_df['Mutation.Type'] = combined_df.apply(mutation_class, axis=1) ## creates a Mutation.type column and stores the outcome
                                                                            ##applying this to all the lines, 1=specifies rows
```

```python
## FIND OUT THE RANGE OF THE MUTATIONS
```

```python
def find_mutation_position_range(row):
    ## specifying the WT and mutant sequences
    wildtype_seq = row['WildType.Sequence']
    mutant_seq = row['Mutant.Sequence']
    
    ## substitutions
    if len(wildtype_seq) == len(mutant_seq):
        diff_positions = []
        for i in range(len(wildtype_seq)):
            if wildtype_seq[i] != mutant_seq[i]:
                diff_positions.append(i + 1)  ## positions are stored in +1 index to account for biological accuracy
        
        ## range
        if diff_positions:
            return (min(diff_positions), max(diff_positions))  
    
    ## insertions
    elif len(wildtype_seq) < len(mutant_seq):  ## specifying insertions by the differences in sequence lengths
        
        for i in range(len(wildtype_seq)):
            if wildtype_seq[i] != mutant_seq[i]:
                return (i + 1, i + len(mutant_seq) - len(wildtype_seq))  ## start position as the first mismatch, stop position by the difference in length of the sequences
        return (len(wildtype_seq) + 1, len(mutant_seq))  ## insertion after the last match
    
    ## deletions
    elif len(wildtype_seq) > len(mutant_seq):  ## specifying insertions by the differences in sequence lengths
        
        for i in range(len(mutant_seq)):
            if wildtype_seq[i] != mutant_seq[i]:
                return (i + 1, i + len(wildtype_seq) - len(mutant_seq))  ## start position as the first mismatch, stop position by the difference in length of the sequences
        return (len(mutant_seq) + 1, len(wildtype_seq))  ## deletion after the last match

    ## if no mutation (wildtype == mutant), return 'No Mutation'
    return 'No Mutation'

## applying to each row
combined_df['Mutation.Position.Range'] = combined_df.apply(find_mutation_position_range, axis=1)


print(combined_df[['Gene', 'Mutation.Position.Range']])
```

```python

## creating new data frame fro cell viability statistics, with columns
cell_viability_stats_df = pd.DataFrame(columns=[
    'Gene', 'WT_CellViability_Mean', 'WT_CellViability_StdDev', 'WT_CellViability_Min', 'WT_CellViability_Max',
    'Mut_CellViability_Mean', 'Mut_CellViability_StdDev', 'Mut_CellViability_Min', 'Mut_CellViability_Max'])

## iterating over each row from the original data frame
for index, row in combined_df.iterrows():
    ## WT data
    wt_cell_viability = [
        row['CellViability.WT.Rep1'],
        row['CellViability.WT.Rep2'],
        row['CellViability.WT.Rep3']]
    
    ## mutatnt data
    mut_cell_viability = [
        row['CellViability.Mut.Rep1'],
        row['CellViability.Mut.Rep2'],
        row['CellViability.Mut.Rep3']]
    
    ## stats calculation using numpy
    ## WT
    wt_mean = np.mean(wt_cell_viability)
    wt_stddev = np.std(wt_cell_viability)
    wt_min = np.min(wt_cell_viability)
    wt_max = np.max(wt_cell_viability)
    
    ## mutant
    mut_mean = np.mean(mut_cell_viability)
    mut_stddev = np.std(mut_cell_viability)
    mut_min = np.min(mut_cell_viability)
    mut_max = np.max(mut_cell_viability)
    
    ## organising the data frame
    new_row = {
        'Gene': row['Gene'],
        'WT_CellViability_Mean': wt_mean,
        'WT_CellViability_StdDev': wt_stddev,
        'WT_CellViability_Min': wt_min,
        'WT_CellViability_Max': wt_max,
        'Mut_CellViability_Mean': mut_mean,
        'Mut_CellViability_StdDev': mut_stddev,
        'Mut_CellViability_Min': mut_min,
        'Mut_CellViability_Max': mut_max}
    
        ## checking for NaN values due to a Pandas recommendation msg
    if not any(pd.isna(value) for value in new_row.values()):
        ## adding the rows to the data frame
       cell_viability_stats_df = pd.concat([cell_viability_stats_df, pd.DataFrame([new_row])], ignore_index=True)
```

```python
## MRNA EXPRESSION
```

```python
## creating new data frame for MRNA expression levels statistics
mRNA_stats_df = pd.DataFrame(columns=['Gene', 
                                      'WT_mRNA_Mean', 'WT_mRNA_StdDev', 'WT_mRNA_Min', 'WT_mRNA_Max', 
                                      'Mut_mRNA_Mean', 'Mut_mRNA_StdDev', 'Mut_mRNA_Min', 'Mut_mRNA_Max',
                                      'PercentChange_mRNA'])

## iterating over each row from the original data frame
for index, row in combined_df.iterrows():
    
    ## extracting mRNA data for WT
    wt_mrna_values = row[['mRNA.Expression.WT.Rep1', 'mRNA.Expression.WT.Rep2', 'mRNA.Expression.WT.Rep3']].values

    ## extracting mRNA data for mutant
    mut_mrna_values = row[['mRNA.Expression.Mut.Rep1', 'mRNA.Expression.Mut.Rep2', 'mRNA.Expression.Mut.Rep3']].values

    ## calculations for WT
    wt_mean = np.mean(wt_mrna_values)
    wt_stddev = np.std(wt_mrna_values)
    wt_min = np.min(wt_mrna_values)
    wt_max = np.max(wt_mrna_values)

    ## calculations for mutant
    mut_mean = np.mean(mut_mrna_values)
    mut_stddev = np.std(mut_mrna_values)
    mut_min = np.min(mut_mrna_values)
    mut_max = np.max(mut_mrna_values)

    ## percent change in mRNA expression
    percent_change_mrna = ((mut_mean - wt_mean) / mut_mean) * 100
    

    ## organising the data frame
    new_row = {
        'Gene': row['Gene'],
        'WT_mRNA_Mean': wt_mean,
        'WT_mRNA_StdDev': wt_stddev,
        'WT_mRNA_Min': wt_min,
        'WT_mRNA_Max': wt_max,
        'Mut_mRNA_Mean': mut_mean,
        'Mut_mRNA_StdDev': mut_stddev,
        'Mut_mRNA_Min': mut_min,
        'Mut_mRNA_Max': mut_max,
        'PercentChange_mRNA': percent_change_mrna}

    ## adding rows to the data frame
    mRNA_stats_df = pd.concat([mRNA_stats_df, pd.DataFrame([new_row])], ignore_index=True)

```

```python
## create new data frame for protein expression statistics
protein_stats_df = pd.DataFrame(columns=['Gene', 'WT_Protein_Expression_Mean', 'Mut_Protein_Expression_Mean', 
                                         'WT_Protein_Expression_StdDev', 'Mut_Protein_Expression_StdDev', 
                                         'PercentChange_Protein'])

## iterating over each row from the original data frame
for index, row in combined_df.iterrows():
    ## specifying WT values 
    wt_protein_values = row[['Protein.Expression.WT.Rep1', 'Protein.Expression.WT.Rep2', 'Protein.Expression.WT.Rep3']].values
    ## specifying mutant values
    mut_protein_values = row[['Protein.Expression.Mut.Rep1', 'Protein.Expression.Mut.Rep2', 'Protein.Expression.Mut.Rep3']].values
    
    ## calculations for WT
    wt_mean = np.mean(wt_protein_values)
    wt_stddev = np.std(wt_protein_values)
    ## calculations for mutant
    mut_mean = np.mean(mut_protein_values)
    mut_stddev = np.std(mut_protein_values)

    ## percent change in protein expression
    percent_change_protein = ((mut_mean - wt_mean) / wt_mean) * 100  

    ## organising the data frame
    new_row = {
        'Gene': row['Gene'],
        'WT_Protein_Expression_Mean': wt_mean,
        'Mut_Protein_Expression_Mean': mut_mean,
        'WT_Protein_Expression_StdDev': wt_stddev,
        'Mut_Protein_Expression_StdDev': mut_stddev,
        'PercentChange_Protein': percent_change_protein}

    ## adding the rows to the data frame
    protein_stats_df = pd.concat([protein_stats_df, pd.DataFrame([new_row])], ignore_index=True)

```
